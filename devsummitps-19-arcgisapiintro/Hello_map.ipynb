```python
from arcgis.gis import GIS
gis=GIS()
```

```python
map=gis.map("Palm Springs, CA")
map
```

```python
map.center=[33.824, -116.538]
```

```python
map.zoom=15
```

```python
map.basemaps
```

```python
map.basemap='streets-navigation-vector'
```

## Search for content

```python
poi_content = gis.content.search("Palm Springs POI", item_type= "Feature Layer", max_items=10)
poi_content
```

```python
from IPython.display import display
for item in poi_content:
    display(item)
```

```python
poi=poi_content[3]
map.add_layer(poi)
```

```python
from arcgis.geocoding import geocode
```

```python
help(geocode)
```

```python
downtownps=geocode("Palm Springs, CA")[0]
map1=gis.map(downtownps)
map1
```

```python
restaurants=geocode(None, downtownps['extent'], category="Thai Food", max_locations=25)
for restaurant in restaurants:
    popup={
        "title": restaurant['attributes']['PlaceName'],
        "content" : "Phone:" + restaurant['attributes']['Phone'] + "<br>" "Address:" +
        restaurant['attributes']['Place_addr']
        
    }
    
    map1.draw(restaurant['location'], popup)
```
